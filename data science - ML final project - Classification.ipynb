### Fine-tune your model and make pipeline
#### Grid Search

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import SGDClassifier
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

logistic = LogisticRegression()

pca = PCA()

pipe = Pipeline([
        ('std_scaler', StandardScaler()),
        ('pca', pca),
        ('logistic', logistic),
    ])

# Create a list of a sequence of integers from 1 to 30 (the number of features in X + 1)
n_components = list(range(1,Hedgefund_predictor.shape[1]+1,1))

# Create a list of values of the regularization parameter
C = np.logspace(-4, 4, 50)

# Create a list of options for the regularization penalty
penalty = ['l1', 'l2']

# Create a dictionary of all the parameter options 
# Note has you can access the parameters of steps of a pipeline by using '__’
param_grid = dict(pca__n_components=n_components, 
                  logistic__C=C,
                  logistic__penalty=penalty)


search = GridSearchCV(pipe, param_grid, iid=False, cv=5,
                      return_train_score=False)

search.fit(train_set, train_label)
print("Best parameter (CV score=%0.3f):" % search.best_score_)
print(search.best_params_)


Pred = search.predict(test_set)

print (" LG Model test Accuracy:: ", accuracy_score(test_label, Pred))
```

### Feature Importance

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

param_distribs = {
        'n_estimators': randint(low=1, high=200),
        'max_features': randint(low=1, high=Hedgefund_predictor.shape[1]),
    }

forest_clf = RandomForestClassifier(random_state=42)
rnd_search = RandomizedSearchCV(forest_clf, param_distributions=param_distribs,
                                n_iter=10, cv=5, scoring='accuracy', random_state=42)
rnd_search.fit(train_set, train_label)
    
feature_importances = rnd_search.best_estimator_.feature_importances_

importance = sorted(zip(feature_importances, Hedgefund_predictor), reverse=True)
importance
```

```python
df_importance = pd.DataFrame(feature_importances, index=Hedgefund_predictor.columns, columns=['Importance'])
df= df_importance.sort_values('Importance', ascending=False)
df.plot(kind='bar', y='Importance', grid=True)
```
