### Select and train a model


```python
from pandas.tools.plotting import scatter_matrix
from sklearn import cross_validation
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn import ensemble
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn import neighbors
from sklearn import ensemble
from sklearn.decomposition import PCA

lin_reg = LinearRegression()

num_folds = 10
num_instances = len(train_set)
seed = 7
scoring = 'neg_mean_squared_error'

models = []

#Linear Algorithms
models.append(('lin_reg', LinearRegression()))

#Non-Linear Algorithms
models.append(('KNN', neighbors.KNeighborsRegressor()))
models.append(('svr', SVR()))

#Emsemble Algorithms
models.append(('GradientBoosting', ensemble.GradientBoostingRegressor()))
models.append(('RANDOM', RandomForestRegressor()))

results = []
names = []

pca = PCA(n_components = Hedgefund_predictor.shape[1])
X_pca = pca.fit_transform(train_set)

for name, model in models:
    kfold = cross_validation.KFold(n=num_instances, n_folds=num_folds, random_state=seed)
    cv_results = cross_validation.cross_val_score(model, X_pca, train_label, cv=kfold, scoring=scoring)
    results.append(-cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, -cv_results.mean(), cv_results.std())
    print(msg)

fig = plt.figure()
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

### Fine-tune your model and make pipeline


#### Grid Search for GradientBoosting

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import SGDClassifier
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error


GradientBoosting = ensemble.GradientBoostingRegressor()

pca = PCA()

pipe = Pipeline([
        ('std_scaler', StandardScaler()),
        ('pca', pca),
        ('GradientBoosting', GradientBoosting),
    ])

# Create a list of a sequence of integers from 1 to the number of features in X + 1
n_components = list(range(1,Hedgefund_predictor.shape[1]+1,1))

n_estimators = [50,100]
learning_rate = [.01,0.1]
max_depth = [2,4]
subsample = [.5,.75]

# Create a dictionary of all the parameter options 
# Note has you can access the parameters of steps of a pipeline by using '__’
param_grid = dict(pca__n_components=n_components, 
                  GradientBoosting__n_estimators=n_estimators,
                  GradientBoosting__learning_rate=learning_rate,
                  GradientBoosting__max_depth=max_depth,
                  GradientBoosting__subsample=subsample
                 )

search = GridSearchCV(pipe, param_grid, iid=False, cv=5, return_train_score=False)

search.fit(train_set, train_label)
print("Best parameter (CV score=%0.3f):" % search.best_score_)
print(search.best_params_)

Pred = search.predict(test_set)

Boosting_mse = mean_squared_error(test_label, Pred)
Boosting_rmse = np.sqrt(Boosting_mse)
print('RSME Based On Gradient Boosting: ', Boosting_rmse)
```

### Feature Importance

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

param_distribs = {
        'n_estimators': randint(low=1, high=200),
        'max_features': randint(low=1, high=Hedgefund_predictor.shape[1]),
    }

forest_clf = RandomForestRegressor(random_state=42)
rnd_search = RandomizedSearchCV(forest_clf, param_distributions=param_distribs,
                                n_iter=10, cv=5, scoring='neg_mean_squared_error', random_state=42)
rnd_search.fit(train_set, train_label)
    
feature_importances = rnd_search.best_estimator_.feature_importances_

importance = sorted(zip(feature_importances, Hedgefund_predictor), reverse=True)
importance
```

```python
df_importance = pd.DataFrame(feature_importances, index=Hedgefund_predictor.columns, columns=['Importance'])
df= df_importance.sort_values('Importance', ascending=False)
df.plot(kind='bar', y='Importance', grid=True)
```
